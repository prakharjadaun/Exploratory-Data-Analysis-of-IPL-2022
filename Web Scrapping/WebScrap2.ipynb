## **Points Table**

```python
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime
```

```python
now = datetime.now()
```

```python
def repeat(num,times):
    temp=[]
    for i in range(times):
        temp.append(num)
    return temp
```

```python
result = requests.get('https://www.firstpost.com/firstcricket/points-table/series/ipl-2022.html')
```

```python
src = result.content
```

```python
soup = BeautifulSoup(src, 'html.parser')
```

```python
table = soup.find_all("table")
Points_table = table[0]
```

```python
teams = Points_table.find_all("tr")
stats = teams[1].find_all("td")
stats[1].text
```

```python
pos = []
team_name = []
played =  []
won = []
lost = []
tied = []
NRR = []
points = []
for i in range(1,11):
    stats = teams[i].find_all("td")
    pos.append(stats[0].text)
    team_name.append(stats[1].text.strip())
    played.append(stats[2].text)
    won.append(stats[3].text)
    lost.append(stats[4].text)
    tied.append(stats[5].text)
    NRR.append(stats[6].text)
    points.append(stats[7].text)
    
```

```python
day = repeat(22,10)
month = repeat(now.month,10)
```

```python
pt =  {
    'Pos': pos, 
    'Team' : team_name, 
    'Plyd' : played, 
    'Won' : won, 
    'Lost' : lost, 
    'Tied' : tied, 
    'NRR' : NRR,
    'Points': points,
    'Day':day,
    'Month':month
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(pt)
```

```python
df
```

```python
df.to_csv('points_table.csv')
```

```python
temp = "points_table.csv"
Main = "Main\PointsTable.csv"

df = pd.concat(map(pd.read_csv,[Main,temp]),ignore_index=True)
df
```

```python
df.to_csv('Main\PointsTable.csv')
```
