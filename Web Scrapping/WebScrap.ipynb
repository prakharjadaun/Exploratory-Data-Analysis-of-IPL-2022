```python
#importing required libraries
import requests
from bs4 import BeautifulSoup
import csv
```

## **Orange Cap**

```python
#getting the url 
result = requests.get('https://www.firstpost.com/firstcricket/orange-cap-holder/series/ipl-2022.html')
```

```python
#getting the content
src = result.content 
```

```python
#creating the BeautifulSoup object
soup = BeautifulSoup(src,'html.parser')
```

```python
#getting all the tables form the webpage
table = soup.find_all("table")
#getting the first table on the webpage
orange_cap_table = table[0]
```

```python
#getting all the rows of the table 
players = orange_cap_table.find_all("tr")
```

```python
#evaluating whether the dats is fetched or not 
stats = players[1].find_all("td")
```

```python
stats[4].text.strip()
```

```python
#storing all the fetched data 
pos = []
player_name = []
team = []
matches = []
runs = []
for i in range(1,13):
    stats = players[i].find_all("td")
    pos.append(stats[0].text)
    player_name.append(stats[1].text.strip())
    team.append(stats[2].text.strip())
    matches.append(stats[3].text)
    runs.append(stats[4].text)
```

```python
#imporing datetime from datetime
from datetime import datetime
```

```python
#repeat func() : which returns a list of repeated numbers 
def repeat(num, times):
    temp = []
    for i in range(times):
        temp.append(num)
    return temp
```

```python
#creating datetime object
now = datetime.now()
#getting the current day and month
day = repeat(22,12)
month = repeat(now.month,12)
```

```python
#creating a dictionary
oc = {
    'Pos':pos,
    'Player':player_name,
    'Team':team,
    'Matches':matches,
    'Runs':runs,
    'Day':day,
    'Month':month
}
```

```python
#importing pandas for dataframe manipulation
import pandas as pd
```

```python
#creat
df = pd.DataFrame(oc)
```

```python
#printing the dataframe
df
```

```python
#converting dataframe to the csv
df.to_csv('orange_cap.csv')
```

```python
temp = "orange_cap.csv"
Main = "Main\OrangeCap.csv"
#appending csv file to the Main csv file
df = pd.concat(map(pd.read_csv,[Main,temp]),ignore_index=True)
df
```

```python
#converting main dataframe to the csv format
df.to_csv("Main\OrangeCap.csv")
```
